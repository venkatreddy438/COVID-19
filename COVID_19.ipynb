```python
from pandas.tseries.offsets import DateOffset
import numpy as np
import pandas as pd
```

```python
#del us_tot
current = (pd.datetime.now()- DateOffset(days=1)).date().strftime("%m-%d-%Y")
last = '01-22-2020'
```

```python
days_tot = (pd.to_datetime(current) - pd.to_datetime(last)).days
```

```python
us_tot = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_daily_reports/' + current + '.csv')[['Province_State','Country_Region','Last_Update','Confirmed','Deaths','Recovered']]
us_tot.columns = ['Province/State','Country/Region','Date','Confirmed','Dead','Recovered']
```

```python
us_tot['Date'] = pd.to_datetime(us_tot['Date']).apply(lambda x: x.date())
```

```python
us_tot.head()
```

```python
us_tot['Active'] = us_tot['Confirmed'] - (us_tot['Recovered'] + us_tot['Dead'])
```

```python
t_conf = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv')
t_conf.head()
```

```python
conf = pd.melt(t_conf, id_vars=["Country/Region", "Province/State", "Lat", "Long"], var_name="Date", value_name="Confirmed").fillna('')
```

```python
t_dead = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv')
t_dead.head()
```

```python
dead = pd.melt(t_dead, id_vars=["Country/Region", "Province/State", "Lat", "Long"], var_name="Date", value_name="Dead").fillna('')
```

```python
reco = pd.melt(t_reco, id_vars=["Country/Region", "Province/State", "Lat", "Long"], var_name="Date", value_name="Recovered").fillna('')
```
